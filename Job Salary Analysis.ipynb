```python
#1.Retrieve the average salary for each company location.
sql="""
select df6.CompanyLocation,avg(df5.Salary) as avg_sal
from df5 
join df6 
on df5.JobID=df6.JobID
group by CompanyLocation
order by avg(df5.Salary) ASC;
"""
avg = pd.read_sql(sql,conn)
avg.head()
```

```python

#2.Find the total number of employees residing in each location.
sql="""
select df6.EmployeeResidence,count(df2.EmploymentType) as totemp
from df2
join df6
on df6.JobID=df2.JobID
group by EmployeeResidence
order by count(df2.EmploymentType) DESC;
"""
totemp = pd.read_sql(sql,conn)
totemp.sample(15)
```

```python
#3.Calculate the average salary across all company sizes.
sql="""
select df6.CompanySize,avg(df5.Salary) as size_avgsal
from df5 
join df6 
on df5.JobID=df6.JobID
group by CompanySize
order by avg(df5.Salary);
"""
size_avgsal = pd.read_sql(sql,conn)
size_avgsal
```

```python
#4.List the highest and lowest paying jobs based on salary.
sql="""
select df1.JobTitle, min(df5.Salary) as Lowest,max(df5.Salary) as Highest 
from df5 
join df1 
on df5.JobID=df1.JobID
group by JobTitle
order by min(df5.Salary),max(df5.Salary)DESC;
"""
Min = pd.read_sql(sql,conn)
Min

```

```python
#5.Find the average salary per year for each company size.
sql="""
select df6.CompanySize,avg(df5.Salary) as avgsal_size , df6.Year 
from df5 
join df6 
on df5.JobID=df6.JobID
group by Year,CompanySize
order by Year ;
"""
avgsal_size = pd.read_sql(sql,conn)
avgsal_size
```

```python
#6.Identify the company location with the highest total salary.
sql="""
select df6.CompanyLocation,sum(df5.Salary) as high_totsal
from df5 
join df6 
on df5.JobID=df6.JobID
group by Year,CompanySize
order by sum(df5.Salary) DESC ;
"""
high_totsal = pd.read_sql(sql,conn)
high_totsal
```

```python
#14.List the Job Titles, Salary, and Currency of jobs that have a Salary greater than $75,000 USD.
sql="""
select df1.JobTitle, df5.SalaryCurrency , df5.Salary as titcursal
from df5
join df1
on df5.JobID=df1.JobID
where Salary>75000
group by JobTitle
order by df5.SalaryCurrency , df5.Salary
"""
titcursal=pd.read_sql(sql,conn)
titcursal
```

```python
#16.Calculate the average Salary in USD for each Employment Type
sql="""
select df2.EmploymentType,avg(df6.SalaryUSD) as empavgsl
from df6
join df2
on df6.JobID=df2.JobID
group by EmploymentType
order by avg(df6.SalaryUSD)
"""
empavgsl = pd.read_sql(sql,conn)
empavgsl
```

```python
#17.Display the Job Titles and Company Sizes for jobs that have a Salary greater than the average Salary for their respective Employment Types.
sql="""
SELECT df1.JobTitle, df6.CompanySize, ROUND(AVG(df6.SalaryUSD)) AS jbcmpavg
FROM df6
JOIN df1 ON df6.JobID = df1.JobID
GROUP BY df1.JobTitle, df6.CompanySize
HAVING AVG(df6.SalaryUSD) > (
    SELECT AVG(SalaryUSD)
    FROM df6
)
ORDER BY AVG(df6.SalaryUSD);
"""
jbcmpavg=pd.read_sql(sql,conn)
jbcmpavg
```

```python
#18.Find the top 5 Company Locations with the highest average Salary in USD.
sql="""
SELECT CompanyLocation, AVG(SalaryUSD) AS top5avvg
FROM df6
GROUP BY CompanyLocation
ORDER BY top5avvg DESC
LIMIT 5
"""
top5avvg=pd.read_sql(sql,conn)
top5avvg
```
